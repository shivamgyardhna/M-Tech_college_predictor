```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import pandas as pd
import time

service = Service("chromedriver.exe")
driver = webdriver.Chrome(service=service)

url = "https://admissions.nic.in/admiss/admissions/orcrjacd/105012421"
driver.get(url)

time.sleep(6)  # wait for Angular + table

rows = driver.find_elements(By.CSS_SELECTOR, "table tr")

data = []
for row in rows[1:]:
    cells = row.find_elements(By.TAG_NAME, "td")
    data.append([cell.text for cell in cells])

columns = [
    "SNo", "Round", "Institute", "PG Program",
    "Group", "Category", "Max GATE Score", "Min GATE Score"
]

df = pd.DataFrame(data, columns=columns)

driver.quit()

print(df.head())
df.to_csv("ccmt_orcr.csv", index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

service = Service("chromedriver.exe")
driver = webdriver.Chrome(service=service)

url = "https://admissions.nic.in/admiss/admissions/orcrjacd/105012421"
driver.get(url)

wait = WebDriverWait(driver, 20)

# Wait for table to appear
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table")))

all_data = []
page = 1

while True:
    print(f"Scraping page {page}")

    # Wait for table rows
    rows = wait.until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "table tbody tr"))
    )

    for row in rows:
        cells = [cell.text for cell in row.find_elements(By.TAG_NAME, "td")]
        all_data.append(cells)

    try:
        # Locate Next button by aria-label
        next_btn = wait.until(
            EC.presence_of_element_located(
                (By.XPATH, "//a[@aria-label='Next page']")
            )
        )

        # Check if disabled (Angular hides clickability)
        if "disabled" in next_btn.get_attribute("class").lower():
            print("Last page reached")
            break

        # Click using JavaScript (IMPORTANT)
        driver.execute_script("arguments[0].click();", next_btn)

        page += 1
        time.sleep(1.2)  # allow Angular to refresh table

    except Exception as e:
        print("No Next page found. Stopping.")
        break

driver.quit()

columns = [
    "SNo", "Round", "Institute", "PG Program",
    "Group", "Category", "Max GATE Score", "Min GATE Score"
]

df = pd.DataFrame(all_data, columns=columns)


print("✅ Done:", df.shape)
```

```python
df.to_csv("ccmt_2024_data.csv", index=False)
```
